### Ý tưởng hình thành:
Với $\mathbf{x} \in X$ and activation function $\sigma$:

$\mathbf{x}_1$ là dữ liệu thứ 1 (vector) trong tập dữ liệu $X$

$x_1$ là đặc trưng thứ 1 (scalar) trong vector $\mathbf{x}$

$\hat{y} = \sigma(1*w_0 + x_1*w_1 + ... + x_m*w_m)$

Rút gọn về dạng tích vô hướng của 2 vector, ta được:
$\hat{y} = \sigma(1*w_0 + \mathbf{w}^\top\mathbf{x})$

Thêm 1 vào đầu vector $\mathbf{x}$ và $w_0$ vào đầu vector $\mathbf{w}$ để có $\mathbf{\dot{x}}$ và $\mathbf{\dot{w}}$, ta được:
$\hat{y} = \sigma(\mathbf{\dot{w}}^\top\mathbf{\dot{x}})$

Trong ví dụ này, activation fucntion là hàm $sign$, ta có

$\hat{y} = \mathtt{sign}(\mathbf{\dot{w}}^{\top} \mathbf{\dot{x}})$

### Giải thuật Perceptron:
**Gradient Descent**:

B1: Chọn $\mathbf{\dot{w}}$ ngẫu nhiên

B2: Tính các $\hat{y}_i$ ứng với điểm dữ liệu $\mathbf{x}_i$. Hoặc có thể nói là tính $\mathbf{\hat{y}}$ ứng với $X$:

> $\hat{y}_i = \mathtt{sign}(\mathbf{\dot{w}}^{\top} \mathbf{\dot{x}}_i)$

B3: Tính loss: Tổng số điểm bị gán nhãn sai.

B4: Tính gradient:
> $\nabla\mathbf{\dot{w}} = -\mathbf{\dot{x}}_i\mathbf{y}_i$

B5: Cập nhật bộ trọng số
> $\mathbf{\dot{w}} = \mathbf{\dot{w}} - \eta\nabla\mathbf{\dot{w}}$ với các điểm $\mathbf{\dot{x}}_i$ bị phân loại sai

**Stochastic Gradient Descent** (SGD):

B1: Chọn $\mathbf{\dot{w}}$ ngẫu nhiên

B2: Tính từng $\hat{y}_i$ ứng với điểm dữ liệu $\mathbf{x}_i$

> $\hat{y}_i = \mathtt{sign}(\mathbf{\dot{w}}^{\top} \mathbf{\dot{x}}_i)$

B3: Tính loss: (Không tính)

*Nếu $\hat{y}_i \neq y_i$ thì thực hiện B4, B5. Nếu không thì lặp lại B2 với điểm dữ liệu tiếp theo*

B4: Tính gradient:
> $\nabla\mathbf{\dot{w}} = -\mathbf{\dot{x}}_i\mathbf{y}_i$

B5: Cập nhật bộ trọng số
> $\mathbf{\dot{w}} = \mathbf{\dot{w}} - \eta\nabla\mathbf{\dot{w}}$

Lặp lại B2 cho lượt huấn luyện (epoch) kế tiếp
