# Basic operations
**Operations between arrays and scalars**:
* Any operations between equal-size arrays applies the operation element-wise
    * Example: $x = \begin{bmatrix} x_1 \\ \vdots \\ x_n \end{bmatrix}$ and $y = \begin{bmatrix} y_1 \\ \vdots \\ y_n \end{bmatrix}$ then 
        * $x + y = \begin{bmatrix} x_1 + y_1 \\ \vdots \\ x_n + y_n \end{bmatrix}$ and $x - y = \begin{bmatrix} x_1 - y_1 \\ \vdots \\ x_n - y_n \end{bmatrix}$
        * $x y = \begin{bmatrix} x_1 y_1 \\ \vdots \\ x_n y_n \end{bmatrix}$ and $x / y = \begin{bmatrix} x_1 / y_1 \\ \vdots \\ x_n / y_n \end{bmatrix}$
        * $x \geq y$ results in $\begin{bmatrix} z_1 \\ \vdots \\ z_n \end{bmatrix}$ where $z_i = \begin{cases} \text{True} & x_i \geq y_i \\ \text{False} & \text{otherwise} \end{cases}$
        * $(x \text{ and } y) = \begin{bmatrix} (x_1 \text{ and } y_1) \\ \vdots \\ (x_n \text{ and } y_n) \end{bmatrix}$
        * $x^y = \begin{bmatrix} x_1^{y_1} \\ \vdots \\ x_n^{y_n} \end{bmatrix}$

# Các phép toán (operation) cơ bản
**Các phép toán giữa mảng và số vô hướng**:
* Các phép toán giữa các mảng có cùng kích thước được thực thi trên từng phần tử tương ứng (element-wise)
    * Ví dụ: $x = \begin{bmatrix} x_1 \\ \vdots \\ x_n \end{bmatrix}$ và $y = \begin{bmatrix} y_1 \\ \vdots \\ y_n \end{bmatrix}$ thì 
        * $x + y = \begin{bmatrix} x_1 + y_1 \\ \vdots \\ x_n + y_n \end{bmatrix}$ và $x - y = \begin{bmatrix} x_1 - y_1 \\ \vdots \\ x_n - y_n \end{bmatrix}$
        * $x y = \begin{bmatrix} x_1 y_1 \\ \vdots \\ x_n y_n \end{bmatrix}$ và $x / y = \begin{bmatrix} x_1 / y_1 \\ \vdots \\ x_n / y_n \end{bmatrix}$
        * $x \geq y$ cho kết quả là $\begin{bmatrix} z_1 \\ \vdots \\ z_n \end{bmatrix}$ với $z_i = \begin{cases} \text{True} & x_i \geq y_i \\ \text{False} & \text{ngược lại} \end{cases}$
        * $(x \text{ và } y) = \begin{bmatrix} (x_1 \text{ và } y_1) \\ \vdots \\ (x_n \text{ và } y_n) \end{bmatrix}$
        * $x^y = \begin{bmatrix} x_1^{y_1} \\ \vdots \\ x_n^{y_n} \end{bmatrix}$

```
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
print(f"addition: {arr1 + arr2}")
print("---")
print(f"compare: {arr1 > arr2}")
```

* Any operations with scalars propagating the scalar to each element
    * Example: $x = \begin{bmatrix} x_1 \\ \vdots \\ x_n \end{bmatrix}$ and $c$ is a scalar then
        * $c x = \begin{bmatrix} c x_1 \\ \vdots \\ c x_n \end{bmatrix}$ and $x / c = \begin{bmatrix} x_1 / c \\ \vdots \\ x_n / c \end{bmatrix}$
        * $x + c = \begin{bmatrix} x_1 + c \\ \vdots \\ x_n + c \end{bmatrix}$ and $x - c = \begin{bmatrix} x_1 - c \\ \vdots \\ x_n - c \end{bmatrix}$
        * $x^c = \begin{bmatrix} x_1^c \\ \vdots \\ x_n^c \end{bmatrix}$
        * $x > c = \begin{bmatrix} z_1 \\ \vdots \\ z_n \end{bmatrix}$ where $z_i = \begin{cases} \text{True} & x_i > c \\ \text{False} & \text{otherwise} \end{cases}$

* Các phép toán với số vô hướng được thực thi trên từng phần tử 
    * Ví dụ: $x = \begin{bmatrix} x_1 \\ \vdots \\ x_n \end{bmatrix}$ và $c$ là một số vô hướng thì
        * $c x = \begin{bmatrix} c x_1 \\ \vdots \\ c x_n \end{bmatrix}$ và $x / c = \begin{bmatrix} x_1 / c \\ \vdots \\ x_n / c \end{bmatrix}$
        * $x + c = \begin{bmatrix} x_1 + c \\ \vdots \\ x_n + c \end{bmatrix}$ và $x - c = \begin{bmatrix} x_1 - c \\ \vdots \\ x_n - c \end{bmatrix}$
        * $x^c = \begin{bmatrix} x_1^c \\ \vdots \\ x_n^c \end{bmatrix}$
        * $x > c = \begin{bmatrix} z_1 \\ \vdots \\ z_n \end{bmatrix}$ với $z_i = \begin{cases} \text{True} & x_i > c \\ \text{False} & \text{ngược lại} \end{cases}$

```
arr = np.array([1, 2, 3])
scalar = 3

print(f"addition: {arr+scalar}")
print("---")
print(f"multiplication: {arr*scalar}")
print("---")
print(f"power: {arr**scalar}")
print("---")
print(f"compare: {arr == scalar}")
```

>**NOTE**: Python keywords `and` and `or` don't work with boolean arrays

**Fast element-wise array functions**: 
* Example: if $x = \begin{bmatrix} x_1 \\ \vdots \\ x_n \end{bmatrix}$ and $y = \begin{bmatrix} y_1 \\ \vdots \\ y_n \end{bmatrix}$ then
    * `np.sqrt(x)` results in \begin{bmatrix} \sqrt{x_1} \\ \vdots \\ \sqrt{x_n} \end{bmatrix}
    * `np.exp(x)` results in \begin{bmatrix} \exp{x_1} \\ \vdots \\ \exp{x_n} \end{bmatrix}
    * `np.max(x, y)` results in \begin{bmatrix} \max(x_1, y_1) \\ \vdots \\ \max(x_n, y_n) \end{bmatrix}

>**LƯU Ý**: Các từ khóa Python `and` và `or` không áp dụng với mảng kiểu Boolean

**Các hàm thao tác trên từng phần tử**: 
* Ví dụ: nếu $x = \begin{bmatrix} x_1 \\ \vdots \\ x_n \end{bmatrix}$ và $y = \begin{bmatrix} y_1 \\ \vdots \\ y_n \end{bmatrix}$ thì
    * `np.sqrt(x)` cho kết quả là \begin{bmatrix} \sqrt{x_1} \\ \vdots \\ \sqrt{x_n} \end{bmatrix}
    * `np.exp(x)` cho kết quả là \begin{bmatrix} \exp{x_1} \\ \vdots \\ \exp{x_n} \end{bmatrix}
    * `np.max(x, y)` cho kết quả là \begin{bmatrix} \max(x_1, y_1) \\ \vdots \\ \max(x_n, y_n) \end{bmatrix}

```
arr1 = np.arange(5)
arr2 = 5 * np.random.rand(5)
print(f"original arrays:")
print(arr1)
print("and")
print(arr2)
print("---")

print(f"square root: {np.sqrt(arr1)}")
print("---")
print(f"exponential root: {np.exp(arr1)}")
print("---")
print(f"max: {np.maximum(arr1, arr2)}")
print("---")
```

# Transposing arrays, swapping axes, and concatenation
* Reshape array: in this example, we transform $\begin{bmatrix} 0 & \cdots & 9 \end{bmatrix}$ into $\begin{bmatrix} 0 & \cdots & 4 \\ 5 & \cdots & 9 \end{bmatrix}$ using `reshape` method

# Chuyển vị (transposing) mảng, tráo (swapping) cột, và ghép chuỗi (concatenation)
* Tái định hình (reshape) mảng: trong ví dụ này, ta chuyển $\begin{bmatrix} 0 & \cdots & 9 \end{bmatrix}$ thành $\begin{bmatrix} 0 & \cdots & 4 \\ 5 & \cdots & 9 \end{bmatrix}$ dùng phương thức `reshape`

```
arr = np.arange(10)
print(f"original shape: {arr.shape}")
print(f"original array: {arr}")

arr = arr.reshape((2, 5))
print(f"new shape: {arr.shape}")
print(f"new array: {arr}")
```

* Transpose array: in this example, we transpose $\begin{bmatrix} 0 & \cdots & 4 \\ 5 & \cdots & 9 \end{bmatrix}$ into $\begin{bmatrix} 0 & 5 \\ \vdots & \vdots \\ 4 & 9 \end{bmatrix}$ using `transpose` method

* Chuyển vị mảng: trong ví dụ này, ta chuyển vị $\begin{bmatrix} 0 & \cdots & 4 \\ 5 & \cdots & 9 \end{bmatrix}$ thành $\begin{bmatrix} 0 & 5 \\ \vdots & \vdots \\ 4 & 9 \end{bmatrix}$ sử dụng phương thức `transpose` 

```
arr = np.arange(10)
arr = arr.reshape((2, 5))
print(f"original array: {arr}")

print(f"transposed array: {arr.T} or {arr.transpose((1, 0))}")

print(f"swap two axes: {arr.swapaxes(1, 0)}")
```

* Example 2: generate random set of points $\{\textbf{x}_i\}_{i=1}^{1000}$ where $\textbf{x}_i = (x^{(1)}_i, x^{(2)}_i)$ from multivariate Gaussian with mean $\begin{bmatrix} 2 \\ 2\end{bmatrix}$ and covariance $\begin{bmatrix} 3 & 0 \\ 0 & 1\end{bmatrix}$

* Ví dụ 2: sinh tập hợp điểm ngẫu nhiên $\{\textbf{x}_i\}_{i=1}^{1000}$ trong đó $\textbf{x}_i = (x^{(1)}_i, x^{(2)}_i)$ với phân phối đa biến Gaussian có trung bình $\begin{bmatrix} 2 \\ 2\end{bmatrix}$ và hợp phương sai $\begin{bmatrix} 3 & 0 \\ 0 & 1\end{bmatrix}$

```
import matplotlib.pyplot as plt
import seaborn as sns

# generate random numbers
# sinh số ngẫu nhiên
means = [[2, 2]]
cov = [[3, 0], [0, 1]]
N = 1000
X0 = np.random.multivariate_normal(means[0], cov, N)

# visualize generated numbers
# vẽ đồ thị các điểm vừa được sinh
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].plot(X0[:, 0], X0[:, 1], "ro", alpha=0.2)
ax[0].set_title("Random 2D points")
ax[0].set_xlabel("X")
ax[0].set_ylabel("Y")

ax[1] = sns.kdeplot(X0[:, 0], X0[:, 1], shade=True)
ax[1].set_title("Underlying Gaussian")
ax[1].set_xlabel("X")
ax[1].set_ylabel("Y")

plt.show()
```

## Unique and other set logic

| Method | Description |

| `intersect1d(x, y)` | Compute the sorted, common elements in 1D arrays `x` and `y` |

## Phép logic tập hợp duy nhất và khác

| `union1d(x, y)` | Tính toán các phần tử hợp nhất, đã được sắp xếp trong mảng 1 chiều `x` và `y` |
| `in1d(x, y)` | Tính toán mảng boolean chỉ ra rằng liệu phần tử `x` có thuộc `y` |
